## 3차 다항식을 활용한 sine 함수 예측
- object: regression
- algorithm: MLP
- metric: MSE

### STEPS
1. 모델 정의
2. 모델 순전파
3. 오차 계산
4. 오차 역전파(가중치 업데이트)
5. 학습 종료

```python
import math
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-math.pi, math.pi, 1000)
y = torch.sin(x)

# 임의의 3차식 계수
## randn(): 정규분포를 따르는 랜덤한 값 반환
a = torch.randn(())
b = torch.randn(())
c = torch.randn(())
d = torch.randn(())

y_pred = a * x**3 + b * x**2 + c + x + d
```

```python
plt.subplot(2,1,1)
plt.title("y_true")
plt.plot(x,y)

plt.subplot(2,1,2)
plt.title("y_pred")
plt.plot(x,y_pred)

plt.show()
```

```python
# 가중치 학습
learning_rate = 1e-6

# training
for epoch in range(2000):
    y_pred = a * x**3 + b * x**2 + c + x + d

    loss = (y_pred - y).pow(2).sum().item() # item(): 실수값으로 반환

    if epoch%100 == 0:
        print(f'epoch{epoch}, loss:{loss}')

    grad_y_pred = 2.0 * (y_pred - y) # 미분값
    grad_a = (grad_y_pred * x ** 3).sum() 
    grad_b = (grad_y_pred * x ** 2).sum() 
    grad_c = (grad_y_pred * x).sum()
    grad_d = grad_y_pred.sum()

    # 가중치 업데이트
    a -= learning_rate * grad_a
    b -= learning_rate * grad_b
    c -= learning_rate * grad_c
    d -= learning_rate * grad_d

# 실제 사인 곡선
plt.subplot(3,1,1)
plt.title('y true')
plt.plot(x,y)

# 예측 사인 곡선
plt.subplot(3,1,2)
plt.title('y pred')
plt.plot(x,y_pred)

plt.show()
```
