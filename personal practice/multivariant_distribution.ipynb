```python
import seaborn as sns
import pandas as pd
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
mu = [2, 3]
cov = [[1, 0], [0, 1]]

rv = sp.stats.multivariate_normal(mu, cov)

xx = np.linspace(0, 4, 120)
yy = np.linspace(1, 5, 150)

XX, YY = np.meshgrid(xx, yy) # 벡터를 행렬로 변환

plt.grid(False)
plt.contourf(XX, YY, rv.pdf(np.dstack([XX, YY])))
```

그래프가 원에 가까울수록 상관계수가 0에 가깝다.

```python
plt.contour(XX, YY, rv.pdf(np.dstack((XX, YY))))
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(XX, YY, rv.pdf(np.dstack((XX, YY))))
```
