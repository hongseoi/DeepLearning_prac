**4. 다음 코드를 실행했을 때, 출력될 값을 순서대로 쓰시오.**


$$
x=
\begin{pmatrix}
\begin{pmatrix}
0 & 1 \\
2 & 3
\end{pmatrix}
&
\begin{pmatrix}
4 & 5 \\
6 & 7
\end{pmatrix} \\
\begin{pmatrix}
8 & 9 \\
10 & 11
\end{pmatrix}
&
\begin{pmatrix}
12 & 13 \\
14 & 15
\end{pmatrix}
\end{pmatrix}
$$
0축은 큰 행렬의 행을 의미하므로 `np.sum(x,axis=0)`은 큰 행렬의 두 행을 더해서
$$
\begin{pmatrix}
\begin{pmatrix}
0 & 1 \\
2 & 3
\end{pmatrix} + 
\begin{pmatrix}
8 & 9 \\
10 & 11
\end{pmatrix}
&
\begin{pmatrix}
4 & 5 \\
6 & 7
\end{pmatrix}+ 
\begin{pmatrix}
12 & 13 \\
14 & 15
\end{pmatrix}
\end{pmatrix}
=
\begin{pmatrix}
\begin{pmatrix}
8 & 10 \\
12 & 14
\end{pmatrix}
&
\begin{pmatrix}
16 & 18 \\
20 & 22
\end{pmatrix}
\end{pmatrix}
$$
입니다.  
1축은 큰 행렬의 열을 의미하므로 `np.sum(x,axis=1)`은 큰 행렬의 두 열을 더해서
$$
\begin{pmatrix}
\begin{pmatrix}
0 & 1 \\
2 & 3
\end{pmatrix} + 
\begin{pmatrix}
4 & 5 \\
6 & 7
\end{pmatrix}
&
\begin{pmatrix}
8 & 9 \\
10 & 11
\end{pmatrix}+ 
\begin{pmatrix}
12 & 13 \\
14 & 15
\end{pmatrix}
\end{pmatrix}
=
\begin{pmatrix}
\begin{pmatrix}
4 & 6 \\
8 & 10
\end{pmatrix}
&
\begin{pmatrix}
20 & 22 \\
24 & 26
\end{pmatrix}
\end{pmatrix}
$$
입니다.  
2축은 작은 행렬의 행을 의미하므로 `np.sum(x,axis=2)`은 작은 행렬의 두 행을 더해서
$$
\begin{pmatrix}
\begin{pmatrix}
0+2 & 1+3 \\
4+6 & 5+7
\end{pmatrix}
&
\begin{pmatrix}
8+10 & 9+11 \\
12+14 & 13+15
\end{pmatrix}
\end{pmatrix}
=
\begin{pmatrix}
\begin{pmatrix}
2 & 4 \\
10 & 12
\end{pmatrix}
&
\begin{pmatrix}
18 & 20 \\
26 & 28
\end{pmatrix}
\end{pmatrix}
$$
입니다.  
3축은 작은 행렬의 열을 의미하므로 `np.sum(x,axis=3)`은 작은 행렬의 두 열을 더해서
$$
\begin{pmatrix}
\begin{pmatrix}
0+1 & 2+3 \\
4+5 & 6+7
\end{pmatrix}
&
\begin{pmatrix}
8+9 & 10+11 \\
12+13 & 14+15
\end{pmatrix}
\end{pmatrix}
=
\begin{pmatrix}
\begin{pmatrix}
1 & 5 \\
9 & 13
\end{pmatrix}
&
\begin{pmatrix}
17 & 21 \\
25 & 29
\end{pmatrix}
\end{pmatrix}
$$
입니다.  

```python
x = np.arange(16).reshape(2,2,2,2)
print(x)
for i in range(4):
    print("="*20)
    print(np.sum(x,axis=i))
```

**5. 싸이킷런은 머신러닝에서 사용하는 대표적인 라이브러리이다. 싸이킷런에서 샘플 사진을 다음과 같이 불러온다.**


**(i) shape과 아미지를 출력하시오.**

```python
from sklearn.datasets import load_sample_image

china = load_sample_image('china.jpg')
print(china.shape)

plt.imshow(china)
plt.axis('off')
plt.show()
```

**(ii) RGB 채널을 분리하시오.**

```python
red=china.copy()
red[:,:,1]=0
red[:,:,2]=0

plt.imshow(red)
plt.axis('off')
plt.show()

green=china.copy()
green[:,:,0]=0
green[:,:,2]=0

plt.imshow(green)
plt.axis('off')
plt.show()

blue=china.copy()
blue[:,:,0]=0
blue[:,:,1]=0

plt.imshow(blue)
plt.axis('off')
plt.show()
```

**(iii) flower.jpg 샘플을 불러온후 shape과 이미지를 출력하시오.**

```python
flower = load_sample_image('flower.jpg')
print(flower.shape)

plt.imshow(flower)
plt.axis('off')
plt.show()
```

**(iv) RGB 채널을 분리하시오.**

```python
red=flower.copy()
red[:,:,1]=0
red[:,:,2]=0

plt.imshow(red)
plt.axis('off')
plt.show()

green=flower.copy()
green[:,:,0]=0
green[:,:,2]=0

plt.imshow(green)
plt.axis('off')
plt.show()

blue=flower.copy()
blue[:,:,0]=0
blue[:,:,1]=0

plt.imshow(blue)
plt.axis('off')
plt.show()
```
