**3. 합성곱층의 필터는 통상적으로 3$\times$3나 5$\times$5를 쓰는 경우가 많다. 합성곱 신경망에서 각 합성곱층의 필터의 해상도를 모두 동일하게 잡더라도 낮은 층의 필터는 좁은 영역의 로컬한 특징을 찾아내지만 높은 층으로 갈수록 점점 더 넓은 영역의  글로벌한 특징을 찾아낸다. 이게 가능한 이유를 설명하시오.**

예를 들어, 합성곱층의 필터를 3$\times$3로 잡고 max pooling층의 필터를 2$\times$2, stride=2로 잡았다고 하죠.  
첫번째 합성곱층에서는 필터가 커버하는 영역이 3$\times$3로 매우 좁습니다.  
max pooling층에서는 2$\times$2 블럭을 픽셀 하나로 바꾸므로 해상도를 절반으로 떨어뜨립니다.  
이때문에 두번째 합성곱층 필터의 크기가 여전히 3$\times$3이지만 최초 입력 이미지에서 커버하는 영역은 6$\times$6로 두배가 됩니다.  
다음 max pooling층에서도 역시 해상도를 절반으로 떨어뜨립니다.  
세번째 합성곱층 필터의 크기는 여전히 3$\times$3이지만 최초 입력 이미지에서 커버하는 영역은 12$\times$12로 네배가 됩니다.

**4. 싸이킷런에서 두 샘플 사진을 다음과 같이 불러오자.**

```python
from sklearn.datasets import load_sample_image

china = load_sample_image('china.jpg')
flower = load_sample_image('flower.jpg')
```

**(i) 두 사진을 묶고 축 변환을 통해 N$\times$C$\times$H$\times$W 포맷의 4차원 텐서를 만드시오.**

```python
images = np.array([china,flower])
images = images.transpose(0,3,1,2)
```

**(ii) PH=3, PW=3 stride=3, padding=0으로 max pooling을 한후 출력 shape N$\times$C$\times$OH$\times$OW를 확인하시오.**

```python
pool_layer = Pooling(3,3,3)
out = pool_layer.forward(images)
print(out.shape)
```

**(iii) max pooling한 결과를 출력하시오.**

```python
import matplotlib.pyplot as plt

for i in (0,1):
    plt.imshow(out[i,:,:,:].transpose(1,2,0)/255)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

**(iv) PH=5, PW=5 stride=5, padding=0에 대해 같은 작업을 하시오.**

```python
pool_layer = Pooling(5,5,5)
out = pool_layer.forward(images)
print(out.shape)

for i in (0,1):
    plt.imshow(out[i,:,:,:].transpose(1,2,0)/255)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```
