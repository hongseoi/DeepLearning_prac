**2. 케라스에서는 합성곱층 인스턴스를 만들때 `padding='same'`으로 설정하면 입력 해상도를 stride로 나눈 값이 출력해상도가 되도록 케라스가 padding 값을 자동으로 잡아준다. 합성곱층으로 들어오는 데이터의 입력 해상도가 $25 \times 25$이고, 필터의 해상도는 $5 \times 5$이라 하자. `padding='same'`, `stride=1`으로 설정했을 때, 케라스가 실제로 잡은 padding값을 구하시오.**

`padding='same'`, `stride=1`으로 설정하면 출력 해상도는 입력 해상도와 같다.  
$$
25 = {W+2P-FW \over S} +1 = {25+2P-5 \over 1} +1
$$
으로부터 $P=2$이다.
