다음과 같은 심층 신경망 Net을 만듭니다.
- 뉴런수는 100 → 100 → 50 → 25 → 10
- 활성화 함수는 디폴트인 ReLu
- 가중치 초기값은 디폴트인 He 초기값

가중치 행렬은 $W_1$, $W_2$, $W_3$, $W_4$ 총 4개가 있고 다음과 같은 분포를 따릅니다.
$$
W_1 \sim N(0,{2 \over 100}), \quad W_2 \sim N(0,{2 \over 100}), \quad W_3 \sim N(0,{2 \over 50}), \quad W_4 \sim N(0,{2 \over 25})
$$
`np.mean(Net.params['W'_str(i)]**2)`는 $i$번째 가중치 행렬의 원소들을 모두 제곱한 후 평균하라는 뜻입니다.  
가중치 행렬의 원소들의 평균이 0이므로 이는 $i$번째 가중치 행렬의 원소들의 분산과 같습니다.
따라서, ${\bf x}$는 가중치 행렬들의 분산의 합
$$
{2 \over 100} + {2 \over 100} + {2 \over 50} + {2 \over 25} = 0.16
$$
과 같습니다.
