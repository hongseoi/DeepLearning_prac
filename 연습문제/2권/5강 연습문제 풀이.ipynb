**1. 왼쪽은 경사하강법으로 오른쪽은 Momemtum으로 이동한 점의 경로를 나타낸 그림이다. 수학(등위선, 접선, gradient, 벡터, 벡터의 덧셈)을 사용하여 점의 이동 경로를 각각 설명하시오.**


경사하강법은 그레디언트의 반대방향으로 이동해 갑니다.  
그레디언트는 등위선에 항상 수직입니다.  
파란선이 등위선의 접선이고 빨간 화살표가 $- \eta \nabla f({\bf x_n})$입니다.

녹색 화살표는 이전 속도에 관성계수를 곱한 $\alpha {\bf v_{n-1}}$입니다.  
노란 화살표는 $- \eta \nabla f({\bf x_n})$입니다.  
녹색 화살표와 노란 화살표를 더해 빨간색 화살표 ${\bf v_n}$를 구하고 이동합니다.  
녹색 화살표 때문에 Momentum에 비해 더 크게 커브를 틀고 더 적은 회수로 목적지에 도달할 수 있습니다.


**2. AdaGrad를 사용하여도 오른쪽과 비슷한 이동 경로를 얻을 수 있다. 수학(편미분)을 사용하여 점의 이동경로를 설명하시오.**

첫번째 빨간 화살표를 보면 세로가 가로 보다 깁니다.  
$y$에 대한 편미분 ${\partial f \over \partial y}$의 크기가 $x$에 대한 편미분 ${\partial f \over \partial x}$의 크기보다 크다는 뜻입니다.  
$$
{\bf h_n} = {\bf h_{n-1}} + \nabla f({\bf x_{n-1}}) \odot \nabla f({\bf x_{n-1}}) = {\bf h_{n-1}} + \left(\left({\partial f \over \partial x}\right)^2, \left({\partial f \over \partial y}\right)^2 \right)
$$
에서 두번째 좌표의 증가폭이 첫번째보다 큽니다.  
노란 화살표는 $- \eta \nabla f({\bf x_n})$입니다.  
$1 \over {\sqrt{\bf h_n}}$을 곱해서 보폭을 조정하면 두번째 좌표의 보폭이 첫번째 좌표의 보폭보다 많이 줄어듭니다.  
따라서, 두번째 빨간 화살표가 더 오른쪽으로 틀어져서 나옵니다.
