**(i) shape  .**

```python
from sklearn.datasets import load_sample_image

china = load_sample_image('china.jpg')
print(china.shape)

plt.imshow(china)
plt.axis('off')
plt.show()
```

**(ii) RGB  .**

```python
red=china.copy()
red[:,:,1]=0
red[:,:,2]=0

plt.imshow(red)
plt.axis('off')
plt.show()

green=china.copy()
green[:,:,0]=0
green[:,:,2]=0

plt.imshow(green)
plt.axis('off')
plt.show()

blue=china.copy()
blue[:,:,0]=0
blue[:,:,1]=0

plt.imshow(blue)
plt.axis('off')
plt.show()
```

```python
flower = load_sample_image('flower.jpg')
print(flower.shape)

plt.imshow(flower)
plt.axis('off')
plt.show()
```
