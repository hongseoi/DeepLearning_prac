**1. 극한, $y$ 절편, 일계미분, 이계미분을 이용하여 시그모이드 함수의 그래프를 그리시오.**

시그모이드 함수는 
$$
y = {1 \over 1+e^{-x}}
$$
으로 정의됩니다.  
극한을 구하면
$$
\lim_{x \to +\infty} {1 \over 1+e^{-x}} = 1, \qquad \lim_{x \to -\infty} {1 \over 1+e^{-x}} = 0
$$
이므로 $y=1$과 $y=0$이 점근선이 됩니다.  
$x=0$을 대입하면 $y$ 절편은
$$
y(0)={1 \over 1+e^{0}} = {1 \over 2}
$$
이 됩니다.  
미분하면
$$
y' = -(1+e^{-x})^{-2} \times (-e^{-x}) = {e^{-x} \over (1+e^{-x})^2}>0
$$
이므로 시그모이드 함수는 증가함수입니다.  
한번 더 미분하면
$$
y'' = {-e^{-x} (1+e^{-x})^2 - e^{-x} \times 2(1+e^{-x}) \times (-e^{-x}) \over (1+e^{-x})^4}
= {e^{-x} (1+e^{-x}) (-1-e^{-x} +2e^{-x}) \over (1+e^{-x})^4}
= {e^{-x} (1+e^{-x}) \over (1+e^{-x})^4} (e^{-x}-1)
$$
입니다.  
부호를 결정하는 항은 $e^{-x}-1$입니다.  
$x$가 0보다 크면 $y''<0$이고 $x$가 0보다 작으면 $y''>0$입니다.  
다시말해, 양수에서는 위로 볼록이고 음수에서는 아래로 볼록입니다.  
그래프를 그리면 다음과 같습니다.

**4. matplotlib을 이용하여 원을 그리시오.**

원 $x^2+y^2=1$은 함수가 아닙니다.  
하나의 $x$에 두개의 $y$값이 대응하죠.  
함수로 이해하기 위해 두 반원으로 나눕니다.  


```python
def semicircle(x):
    return np.sqrt(1-x**2)

plt.figure(figsize=(5,5))
X = np.arange(-1.0, 1.0001, 0.0001)
Y = semicircle(X)
plt.plot(X, Y, 'r', X, -Y, 'r')
plt.show()
```

**7. (i) 시그모이드 함수는 등식
$$
y'=y(1-y)
$$
을 만족함을 보이시오.**

$$
y={1 \over 1+e^{-x}} = (1+e^{-x})^{-1}
$$

$$
\begin{aligned}
y'&=-(1+e^{-x})^{-2} (-e^{-x}) \\
&={e^{-x} \over (1+e^{-x})^2} \\
&={(1+e^{-x})-1 \over (1+e^{-x})^2} \\
&={1 \over 1+e^{-x}}-{1 \over (1+e^{-x})^2} \\
&= y-y^2 \\
&= y(1-y)
\end{aligned}
$$
인공신경망이 데이터로부터 학습하는 수학적 원리는 미분입니다. 따라서, 인공신경망이 빨리 학습하려면 미분을 빨리 계산해야만 합니다. 위 식은 미분을 사칙연산만으로 계산할 수 있게 해줍니다.

**(ii) (i)을 이용하여 시그모이드 함수는 증가 함수임을 보이시오.**

$$
y={1 \over 1+e^{-x}}
$$
로부터
$$
0<y<1
$$
을 알수 있습니다.
(i)에서
$$
y'=y(1-y)
$$
이므로
$$
y'>0
$$
이 됩니다.

**(iii) (i)과 (ii)를 이용하여 시그모이드 함수의 그래프는 $x>0$일때 위로 볼록하고 $x<0$일때 아래로 볼록함을 보이시오.**

(i)에서 얻은
$$
y'=y(1-y)
$$
을 한번 더 미분하면
$$
y'' = y'(1-y) + y(-y') = y' -2yy' = y'(1-2y)
$$
을 얻습니다.
(ii)에서 $y'>0$이므로 $y''$의 부호는 $1-2y$와 같습니다.
$x=0$일때 $y={1 \over 2}$, 즉 $1-2y=0$입니다.
$y$는 증가하므로 $x>0$일때 $y''<0$, 즉 위로 볼록하고 $x<0$일때 $y''>0$, 즉 아래로 볼록합니다.

**8. 1학년 미적분학 시간에 배운 테일러 전개를 시각화 해보자. 함수 $f$에 대하여 다항식
$$
\sum_{k=0}^n {f^{(k)}(0) \over k!} x^k
$$
을 $f$의 $n$차 근사다항식이라 한다. 여기서 $f^{(k)}$는 $f$의 $k$차 도함수이다.**

**(i) $f(x)=\sin x$의 $n$차 근사다항식을 구하시오.**

$\sin x$를 계속 미분해 나가면 다음과 같은 4개의 마디가 반복됩니다.  
$x=0$을 대입하면 $k$가 짝수일 때는 모두 $0$이고 홀수일때는 $1$과 $-1$이 번갈아 가며 등장합니다.  

$\sin x$의 $n$차 근사 다항식은
$$
x - {x^3 \over 3!} + {x^5 \over 5!} - {x^7 \over 7!} + {x^9 \over 9!} - {x^{11} \over 11!} + {x^{13} \over 13!} - {x^{15} \over 15!} + \cdots
$$
입니다.

**(ii) 인터넷으로  numpy의 팩토리얼 함수를 검색하시오.**

np.math.factorial

**(iii) 반복문을 이용하여 (i)을 코드로 구현하시오.**

```python
def P(n,x):
    S=0
    for k in range(1,n+1):
        if k%4==1:
            S+=(x**k)/np.math.factorial(k)
        if k%4==3:
            S-=(x**k)/np.math.factorial(k)
    return S
```

**(iv) matplotlib을 이용하여 $f(x)=\sin x$와 근사다항식들의 그래프를 그리시오.**

1학년 미적분학 시간에 배우는 근사다항식을 matplotlib을 이용해 시각화해봤습니다.  
다항함수의 차수가 커지면서 $\sin x$를 근사하는 영역이 점점 더 넓어집니다.

```python
plt.figure(figsize=(20,10))

X=np.arange(-10,10,0.01)
Y=np.sin(X)
plt.plot(X,Y)

n=15
for i in range(1,n+1):
    plt.plot(X,P(i,X))
plt.ylim(-5,5)
plt.show()
```
